```python
import numpy as np
import random
from tabulate import tabulate
import matplotlib.pyplot as plt

# Constants
GENERATOR_COUNT = 6
MIN_POWER = [1600, 600, 1200, 900, 1800, 290]
MAX_POWER = [3400, 1400, 2100, 2100, 4300, 800]
HEAT_RATE_A = np.array([275.1741, 624.9346, 348.5491, 538.3751, -212.5033, -119.3939])
HEAT_RATE_B = np.array([34.2155, 178.8621, 27.2541, 52.8313, 151.7804, 45.6393])
HEAT_RATE_C = np.array([-0.0015, -0.0512, -0.0019, -0.0003, -0.0083, -0.0061])

# Generator Names
GENERATOR_NAMES = ["Suralaya", "MuaraTawar", "Tanjung Jati", "Gresik", "Paiton", "Grati"]

# Genetic Algorithm Parameters
POPULATION_SIZE = 10
GENERATION_COUNT = 10  # Maximum number of iterations
CROSSOVER_PROBABILITY = 0.8
MUTATION_PROBABILITY = 0.1
EXCHANGE_RATE = 1580  # Rupiah per unit

def plot_fitness_convergence(best_fitness_values):
    generations = range(1, len(best_fitness_values) + 1)
    plt.plot(generations, best_fitness_values, marker='o')
    plt.title('Convergence of Fitness Over Generations')
    plt.xlabel('Generation')
    plt.ylabel('Best Fitness Value')
    plt.show()

def plot_population_distribution(fitness_values, generation_number):
    plt.hist(fitness_values, bins=20, edgecolor='black', alpha=0.7)
    plt.title(f'Distribution of Fitness in Generation {generation_number}')
    plt.xlabel('Fitness Value')
    plt.ylabel('Frequency')
    plt.show()

def plot_optimal_solution(optimal_solution):
    bar_width = 0.3  # You can adjust this value to change the width of the bars
    plt.bar(GENERATOR_NAMES, optimal_solution, width=bar_width)
    plt.title('Optimal Solution for Each Generator using GA')
    plt.xlabel('Generator')
    plt.ylabel('Power Output')
    plt.show()

def calculate_power_demand(generator_powers):
    # Assuming the generator powers represent the power contribution in Megawatts (MW)
    # Calculate the total power demand for each generator
    total_power_demand = generator_powers
    return total_power_demand

def generate_population(population_size, generator_count):
    return np.array([np.random.uniform(low=MIN_POWER, high=MAX_POWER, size=generator_count) for _ in range(population_size)])

def calculate_fitness(solution):
    total_cost = np.sum(HEAT_RATE_A * solution**2 + HEAT_RATE_B * solution + HEAT_RATE_C)
    return total_cost * EXCHANGE_RATE

def crossover(parent1, parent2):
    crossover_point = random.randint(0, GENERATOR_COUNT)  # Include all elements in crossover
    child = np.concatenate((parent1[:crossover_point], parent2[crossover_point:]))
    return child

def mutate(solution):
    mutation_point = random.randint(0, GENERATOR_COUNT - 1)
    solution[mutation_point] = random.uniform(MIN_POWER[mutation_point], MAX_POWER[mutation_point])
    return solution

def calculate_individual_cost(solution):
    """Menghitung biaya optimal untuk setiap pembangkit berdasarkan solusi."""
    individual_costs = HEAT_RATE_A * solution**2 + HEAT_RATE_B * solution + HEAT_RATE_C
    return individual_costs * EXCHANGE_RATE

def display_individual_costs(optimal_solution):
    """Menampilkan biaya optimal untuk setiap pembangkit dalam tabel."""
    individual_costs = calculate_individual_cost(optimal_solution)

    # Membuat tabel untuk menampilkan biaya setiap pembangkit
    cost_table = [[GENERATOR_NAMES[i], f"{individual_costs[i]:.5e}"] for i in range(GENERATOR_COUNT)]
    
    print("\nBiaya Optimal Setiap Pembangkit:")
    print(tabulate(cost_table, headers=["Generator", "Biaya Optimal (Rupiah)"], tablefmt="grid"))

# Tambahkan ini ke dalam fungsi genetic_algorithm setelah solusi optimal ditemukan:
def genetic_algorithm():
    population = generate_population(POPULATION_SIZE, GENERATOR_COUNT)

    # List untuk menyimpan nilai fitness terbaik setiap generasi
    best_fitness_values = []

    for generation in range(GENERATION_COUNT):
        fitness_values = [calculate_fitness(ind) for ind in population]
        sorted_indices = np.argsort(fitness_values)
        sorted_population = population[sorted_indices]

        # Menambahkan nilai fitness terbaik ke list
        best_fitness_values.append(fitness_values[sorted_indices[0]])

        # Menampilkan distribusi populasi setiap generasi
        plot_population_distribution(fitness_values, generation + 1)

        # Display population values
        print(f"Generation: {generation + 1}, Population:")
        print(tabulate(sorted_population, headers=GENERATOR_NAMES, tablefmt="grid"))
        print()  # Move to the next line

        # Display the cost for each individual in the population with generator titles
        print("Cost for each individual:")
        cost_table = [[generator, fitness_values[i]] for i, generator in enumerate(GENERATOR_NAMES)]
        print(tabulate(cost_table, headers=["Generator", "Cost"], tablefmt="grid"))
        print()  # Move to the next line

        # Perform selection (choose parents)
        selected_indices = random.choices(range(POPULATION_SIZE), k=POPULATION_SIZE)
        selected_population = sorted_population[selected_indices]

        new_population = []

        for i in range(0, POPULATION_SIZE, 2):
            parent1, parent2 = selected_population[i], selected_population[i + 1]

            # Perform crossover
            if random.random() < CROSSOVER_PROBABILITY:
                child = crossover(parent1, parent2)
            else:
                child = parent1

            # Perform mutation
            if random.random() < MUTATION_PROBABILITY:
                child = mutate(child)

            new_population.extend([parent1, parent2, child])

        population = np.array(new_population[:POPULATION_SIZE])

    optimal_solution = population[0]
    optimal_cost = calculate_fitness(optimal_solution)

    # Display the optimal solution in the form of a table
    optimal_table = [[generator, optimal_solution[i]] for i, generator in enumerate(GENERATOR_NAMES)]
    print("\nOptimal Solution:")
    print(tabulate(optimal_table, headers=["Generator", "Power"], tablefmt="grid"))
    print("Total cost:", optimal_cost, "Rupiah")

    # Menampilkan grafik konvergensi fitness
    plot_fitness_convergence(best_fitness_values)

    # Menampilkan optimal solution
    optimal_solution = population[0]
    plot_optimal_solution(optimal_solution)

    # Menampilkan solusi optimal dalam tabel setelah grafik
    optimal_solution_table = [[GENERATOR_NAMES[i], optimal_solution[i]] for i in range(GENERATOR_COUNT)]
    print("\nOptimal Solution in Table Format:")
    print(tabulate(optimal_solution_table, headers=["Generator", "Power Output (MW)"], tablefmt="grid"))

    # Tambahkan kode untuk menampilkan biaya individual optimal setiap pembangkit
    print(f"\nTotal biaya optimal: {optimal_cost:.5e} Rupiah")
    display_individual_costs(optimal_solution)

if __name__ == "__main__":
    genetic_algorithm()
```
